```python
import imaplib
import os

# ================== CONFIGURATION ==================
IMAP_SERVER = "imap.mail.yahoo.com"
EMAIL_ACCOUNT = os.getenv("YAHOO_EMAIL") or "degreatmind@yahoo.com"
APP_PASSWORD = os.getenv("YAHOO_APP_PASSWORD") or "dxkgvccwohwgvkdk"
DOMAIN = "linkedin.com"
# ===================================================


def delete_emails_from_sender():
    try:
        # Connect to Yahoo IMAP server
        mail = imaplib.IMAP4_SSL(IMAP_SERVER)

        # Login
        mail.login(EMAIL_ACCOUNT, APP_PASSWORD)
        print("‚úÖ Logged in successfully")

        # Select inbox
        mail.select("INBOX")

        # Search for emails from the specified sender
        status, messages = mail.search(None, f'FROM "{DOMAIN}"')

        if status != "OK":
            print("‚ùå Error searching emails")
            return

        email_ids = messages[0].split()

        if not email_ids:
            print("üì≠ No emails found from this sender")
            return

        print(f"üîé Found {len(email_ids)} emails. Deleting...")

        # Mark emails as deleted
        for email_id in email_ids:
            mail.store(email_id, "+FLAGS", "\\Deleted")

        # Permanently delete
        mail.expunge()

        print("üóë Emails deleted successfully")

        # Logout
        mail.logout()

    except Exception as e:
        print("‚ùå Error:", str(e))


if __name__ == "__main__":
    delete_emails_from_sender()
```
